```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
x = np.linspace(-5,5,2000)
y = sigmoid(x)
```

```python
plt.plot(x,y)
plt.title('Sigmoid')
```

```python
data = pd.read_csv('C:/Users/toduc/OneDrive - National Economics University/Desktop/NEU/Machine Learning/Visualization/Data/fb_stock_prices_2018.csv')
data
```

```python
data['date'] = pd.to_datetime(data['date'])
week = data['date'].dt.week
data['week'] = week
data
```

```python
df_min = data.groupby(['week'])['low'].min()
df_max = data.groupby(['week'])['high'].max()
diff = df_max - df_min
print(df_min, '\n')
print(df_max, '\n')
print(diff)
```

```python
plt.plot(diff)
plt.title('different in stock price of Facebook')
plt.xlabel('week')
plt.ylabel('changes')
```

```python
high = data['high']
low = data['low']
high_low_diff = high - low
plt.scatter(data['volume'], high_low_diff, alpha=0.25, color='black')
plt.title('Relationship between the volumne and different between high and low price')
```
